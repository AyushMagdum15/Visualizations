```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#detting the data
df=pd.read_csv('BAC.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.head()
```

```python
#plotting a line plot of our data
plt.plot(df['Date'],df['Volume'],color='red')
plt.xlabel('Date')
plt.ylabel('Volume')
```

```python
#plotting heatmap to check correlation between data
sns.heatmap(df.corr(), annot=True)
```

```python
#plotting line plot of closing value of each date
plt.figure(figsize=(14,5))
sns.set_style('ticks')
sns.lineplot(data=df,x='Date',y='Close',color='#95230b')
sns.despine()
```

```python
#plotting a pair plot
sns.pairplot(data=df)
```

```python
df.head()
```

```python
#taking more than 2 values of data for comparison and plotting it in a same line plot
plt.figure(figsize=(14,5))
sns.set_style('ticks')
sns.lineplot(data=df,x='Date',y='Open',color='#95230b')
sns.lineplot(data=df,x='Date',y='High',color='#e6ea10')
sns.lineplot(data=df,x='Date',y='Low',color='#0d2aac')
sns.despine()
```

```python
#plotting a histogram
sns.histplot(data=df,x='Volume',bins=10)
```

```python
#creating a
df['year']=df['Date'].dt.year
```

```python
sns.barplot(x='year',y='Volume',data=df)
```
